## Simple Explanation of Bayesian Networks

Bayesian Networks are probabilistic graphical models that represent relationships between variables using directed graphs (like flowcharts with arrows). Think of them as "smart decision trees" that capture how different factors influence each other using probability. Each node represents a variable (like "Rain," "Traffic," or "Late to Work"), and arrows show direct influences. The network uses Bayes' theorem to calculate the probability of events given evidence. For example, if you know it's raining, the network can predict how likely you are to be late to work by considering intermediate factors like traffic. They're powerful for reasoning under uncertainty, making predictions, and understanding cause-and-effect relationships in complex systems.

## Detailed Learning Roadmap

### 1. **Graph Theory Foundations**
   - **Directed Acyclic Graphs (DAGs)**: Learn about nodes, edges, and why cycles aren't allowed
   - **Example**: Family tree showing parent-child relationships (parents → children, no loops)

### 2. **Probability Fundamentals**
   - **Joint, Marginal, and Conditional Probability**: Understanding P(A,B), P(A), and P(A|B)
   - **Example**: P(Rain, Traffic) vs P(Rain) vs P(Traffic|Rain)

### 3. **Conditional Independence**
   - **D-separation rules**: When variables are independent given evidence
   - **Example**: "Wet grass" depends on "Rain" and "Sprinkler," but Rain and Sprinkler are independent

### 4. **Network Construction**
   - **Structure Learning**: Building the graph from data or expert knowledge
   - **Example**: Medical diagnosis network connecting symptoms to diseases

### 5. **Parameter Learning**
   - **Conditional Probability Tables (CPTs)**: Learning probabilities for each node
   - **Example**: P(Fever=High|Disease=Flu) = 0.8

### 6. **Inference Methods**
   - **Exact Inference**: Variable elimination, junction tree algorithms
   - **Example**: Calculating P(Disease|Symptoms) using all available evidence

### 7. **Approximate Inference**
   - **Sampling Methods**: Monte Carlo, Gibbs sampling
   - **Example**: Estimating probabilities when exact calculation is too complex

### 8. **Dynamic Bayesian Networks**
   - **Temporal Modeling**: Adding time dimension to capture sequences
   - **Example**: Stock price prediction considering previous days' data

## FORMULA MEMORY AIDS SECTION

### 1. Bayes' Theorem
**FORMULA**: P(A|B) = P(B|A) × P(A) / P(B)

**REAL-LIFE ANALOGY**: "Medical Test Accuracy"
- P(A|B) = Probability you have the disease given a positive test (what you want to know)
- P(B|A) = Test accuracy - probability of positive test given you have disease (95%)
- P(A) = How common the disease is in population (1 in 1000 people)
- P(B) = How often anyone tests positive (including false positives)

**MEMORY TRICK**: "Bayes flips the question! From 'test accuracy' to 'do I really have it?'"

### 2. Chain Rule of Probability
**FORMULA**: P(X₁, X₂, ..., Xₙ) = ∏ᵢ P(Xᵢ|Parents(Xᵢ))

**REAL-LIFE ANALOGY**: "Recipe Steps Dependencies"
- X₁ = "Buy ingredients" (depends on nothing: P(X₁))
- X₂ = "Preheat oven" (depends on having ingredients: P(X₂|X₁))
- X₃ = "Bake cake" (depends on oven being hot: P(X₃|X₂))
- Total probability = P(X₁) × P(X₂|X₁) × P(X₃|X₂)

**MEMORY TRICK**: "Chain rule = Chain of cooking steps - each step depends on previous ones!"

### 3. Conditional Independence
**FORMULA**: P(X|Y,Z) = P(X|Z) if X ⊥ Y | Z

**REAL-LIFE ANALOGY**: "Weather and Clothing Choice"
- X = "Wearing jacket" 
- Y = "Season (Winter/Summer)"
- Z = "Today's temperature"
- Given today's temperature, the season doesn't matter for jacket choice

**MEMORY TRICK**: "Given the 'middleman' (Z), the other factor (Y) becomes irrelevant!"

## Step-by-Step Numerical Example

Let's build a simple medical diagnosis network:

**Variables**: 
- Flu (F): Yes/No
- Fever (Fv): High/Normal  
- Cough (C): Yes/No

**Network Structure**: Flu → Fever, Flu → Cough

**Step 1: Prior Probabilities**
- P(Flu = Yes) = 0.1
- P(Flu = No) = 0.9

**Step 2: Conditional Probability Tables**
- P(Fever = High | Flu = Yes) = 0.8
- P(Fever = High | Flu = No) = 0.1
- P(Cough = Yes | Flu = Yes) = 0.7
- P(Cough = Yes | Flu = No) = 0.2

**Step 3: Calculate Joint Probability**
P(Flu=Yes, Fever=High, Cough=Yes) = P(Flu=Yes) × P(Fever=High|Flu=Yes) × P(Cough=Yes|Flu=Yes)
= 0.1 × 0.8 × 0.7 = 0.056

**Step 4: Inference with Evidence**
Given evidence (Fever=High, Cough=Yes), calculate P(Flu=Yes|Evidence):

P(Flu=Yes|Fv=High,C=Yes) = P(Fv=High,C=Yes|Flu=Yes) × P(Flu=Yes) / P(Fv=High,C=Yes)

Numerator = 0.8 × 0.7 × 0.1 = 0.056
Denominator = 0.056 + P(Fv=High,C=Yes|Flu=No) × P(Flu=No)
= 0.056 + (0.1 × 0.2 × 0.9) = 0.056 + 0.018 = 0.074

Result: P(Flu=Yes|Evidence) = 0.056/0.074 = 0.757 (75.7%)

```python
# Installation commands for Google Colab and local Mac
# !pip install pgmpy pandas numpy matplotlib seaborn scikit-learn

import logging
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
from pgmpy.estimators import MaximumLikelihoodEstimator, BayesianEstimator
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
import json
import pickle
import warnings
warnings.filterwarnings('ignore')

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('bayesian_network_training.log'),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

def load_and_prepare_data():
    logger.info("Loading and preparing Iris dataset for Bayesian Network")
    
    # Load Iris dataset
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    
    logger.info(f"Original dataset shape: {df.shape}")
    logger.info(f"Features: {list(df.columns)}")
    
    # Discretize continuous features into categories
    discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
    
    # Create meaningful feature names
    feature_mapping = {
        'sepal length (cm)': 'Sepal_Length',
        'sepal width (cm)': 'Sepal_Width', 
        'petal length (cm)': 'Petal_Length',
        'petal width (cm)': 'Petal_Width'
    }
    
    df_processed = df.copy()
    for old_name, new_name in feature_mapping.items():
        discretized_feature = discretizer.fit_transform(df[[old_name]]).flatten()
        df_processed[new_name] = discretized_feature.astype(int)
        logger.info(f"Discretized {old_name} -> {new_name}: {len(np.unique(discretized_feature))} bins")
    
    # Create final dataset with meaningful names
    final_df = df_processed[['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'species']].copy()
    final_df.rename(columns={'species': 'Species'}, inplace=True)
    
    logger.info(f"Final processed dataset shape: {final_df.shape}")
    logger.info(f"Species distribution: {final_df['Species'].value_counts().to_dict()}")
    
    return final_df

def create_network_structure():
    logger.info("Creating Bayesian Network structure")
    
    # Define network structure based on botanical knowledge
    # Species influences all physical characteristics
    model = DiscreteBayesianNetwork([
        ('Species', 'Sepal_Length'),
        ('Species', 'Sepal_Width'),
        ('Species', 'Petal_Length'),
        ('Species', 'Petal_Width'),
        ('Petal_Length', 'Petal_Width')  # Petal dimensions are correlated
    ])
    
    logger.info(f"Network nodes: {model.nodes()}")
    logger.info(f"Network edges: {model.edges()}")
    logger.info("Network structure created successfully (DAG validation will occur after parameter learning)")
    
    return model

def learn_parameters(model, data):
    logger.info("Learning network parameters using Maximum Likelihood Estimation")
    
    # Fit the model to data
    model.fit(data, estimator=MaximumLikelihoodEstimator)
    
    # Now we can check if the model is valid
    try:
        model.check_model()
        logger.info("Model validation passed - all CPDs are properly defined")
    except Exception as e:
        logger.warning(f"Model validation warning: {e}")
    
    logger.info("Learned Conditional Probability Distributions:")
    for cpd in model.get_cpds():
        logger.info(f"CPD for {cpd.variable}:")
        logger.info(f"  Variables: {cpd.variables}")
        
        # Use the correct attribute name for cardinalities
        try:
            if hasattr(cpd, 'cardinality'):
                logger.info(f"  Cardinality: {cpd.cardinality}")
            elif hasattr(cpd, 'get_cardinality'):
                logger.info(f"  Cardinality: {cpd.get_cardinality()}")
            else:
                logger.info(f"  Shape info: {cpd.values.shape}")
        except Exception as e:
            logger.info(f"  Shape info: {cpd.values.shape}")
        
        logger.info(f"  Values shape: {cpd.values.shape}")
        
        # Log a sample of the probability values
        if cpd.values.size <= 20:  # Only for small CPDs
            logger.info(f"  Sample values: {cpd.values.flatten()[:5]}")
    
    return model

def perform_inference(model, evidence_scenarios):
    logger.info("Performing probabilistic inference")
    
    # Create inference object
    inference = VariableElimination(model)
    results = {}
    
    for scenario_name, evidence in evidence_scenarios.items():
        logger.info(f"Inference scenario: {scenario_name}")
        logger.info(f"Evidence: {evidence}")
        
        try:
            # Query probability of each species given evidence
            result = inference.query(variables=['Species'], evidence=evidence)
            
            logger.info(f"Results for {scenario_name}:")
            for i, prob in enumerate(result.values):
                species_name = ['Setosa', 'Versicolor', 'Virginica'][i]
                logger.info(f"  P(Species={species_name}|Evidence) = {prob:.4f}")
            
            results[scenario_name] = {
                'evidence': evidence,
                'probabilities': result.values.tolist(),
                'species_names': ['Setosa', 'Versicolor', 'Virginica']
            }
        except Exception as e:
            logger.error(f"Error in inference for {scenario_name}: {e}")
            results[scenario_name] = {
                'evidence': evidence,
                'error': str(e)
            }
    
    return results

def visualize_network_and_results(model, inference_results, data):
    logger.info("Creating visualizations")
    
    # Set up the plotting style
    plt.style.use('default')
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot 1: Data distribution
    species_counts = data['Species'].value_counts().sort_index()
    axes[0, 0].bar(species_counts.index, species_counts.values, alpha=0.7, edgecolor='black')
    axes[0, 0].set_title('Species Distribution in Dataset')
    axes[0, 0].set_xlabel('Species (0=Setosa, 1=Versicolor, 2=Virginica)')
    axes[0, 0].set_ylabel('Count')
    axes[0, 0].set_xticks([0, 1, 2])
    
    # Plot 2: Feature correlation heatmap
    correlation_matrix = data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, ax=axes[0, 1], fmt='.3f')
    axes[0, 1].set_title('Feature Correlation Matrix')
    
    # Plot 3: Inference results comparison
    valid_results = {k: v for k, v in inference_results.items() if 'probabilities' in v}
    if valid_results:
        scenarios = list(valid_results.keys())[:3]  # Plot first 3 valid scenarios
        species_names = ['Setosa', 'Versicolor', 'Virginica']
        x = np.arange(len(species_names))
        width = 0.25
        
        colors = ['skyblue', 'lightcoral', 'lightgreen']
        for i, scenario in enumerate(scenarios):
            probs = valid_results[scenario]['probabilities']
            axes[1, 0].bar(x + i*width, probs, width, label=scenario, alpha=0.8, color=colors[i])
        
        axes[1, 0].set_xlabel('Species')
        axes[1, 0].set_ylabel('Probability')
        axes[1, 0].set_title('Inference Results Comparison')
        axes[1, 0].set_xticks(x + width)
        axes[1, 0].set_xticklabels(species_names)
        axes[1, 0].legend()
        axes[1, 0].set_ylim(0, 1)
    else:
        axes[1, 0].text(0.5, 0.5, 'No valid inference results', 
                       ha='center', va='center', transform=axes[1, 0].transAxes)
        axes[1, 0].set_title('Inference Results (No Valid Results)')
    
    # Plot 4: Model structure visualization (simplified)
    # Create adjacency matrix for network structure
    nodes = list(model.nodes())
    adj_matrix = np.zeros((len(nodes), len(nodes)))
    node_to_idx = {node: i for i, node in enumerate(nodes)}
    
    for edge in model.edges():
        i, j = node_to_idx[edge[0]], node_to_idx[edge[1]]
        adj_matrix[i][j] = 1
    
    sns.heatmap(adj_matrix, annot=True, xticklabels=nodes, yticklabels=nodes, 
                cmap='Blues', ax=axes[1, 1], cbar=False, fmt='g')
    axes[1, 1].set_title('Network Structure (Adjacency Matrix)')
    axes[1, 1].set_xlabel('Child Nodes')
    axes[1, 1].set_ylabel('Parent Nodes')
    
    plt.tight_layout()
    plt.savefig('bayesian_network_analysis.png', dpi=300, bbox_inches='tight')
    logger.info("Saved visualization as 'bayesian_network_analysis.png'")
    plt.show()

def evaluate_model_performance(model, data):
    logger.info("Evaluating model performance")
    
    # Split data for evaluation
    train_data, test_data = train_test_split(data, test_size=0.3, random_state=42, stratify=data['Species'])
    logger.info(f"Training data shape: {train_data.shape}")
    logger.info(f"Test data shape: {test_data.shape}")
    
    # Train model on training data
    train_model = DiscreteBayesianNetwork(model.edges())
    train_model.fit(train_data, estimator=MaximumLikelihoodEstimator)
    
    # Perform inference on test data
    inference = VariableElimination(train_model)
    correct_predictions = 0
    total_predictions = 0
    
    prediction_details = []
    
    for idx, row in test_data.iterrows():
        try:
            # Use all features except Species as evidence
            evidence = {col: int(row[col]) for col in data.columns if col != 'Species'}
            
            # Query for Species
            result = inference.query(variables=['Species'], evidence=evidence)
            predicted_species = np.argmax(result.values)
            actual_species = int(row['Species'])
            
            prediction_details.append({
                'actual': actual_species,
                'predicted': predicted_species,
                'confidence': float(np.max(result.values)),
                'evidence': evidence
            })
            
            if predicted_species == actual_species:
                correct_predictions += 1
            total_predictions += 1
        except Exception as e:
            logger.warning(f"Could not make prediction for row {idx}: {e}")
    
    if total_predictions > 0:
        accuracy = correct_predictions / total_predictions
        logger.info(f"Model accuracy on test data: {accuracy:.4f} ({correct_predictions}/{total_predictions})")
        
        # Log some prediction examples
        logger.info("Sample predictions:")
        for i, pred in enumerate(prediction_details[:5]):
            result = "✓" if pred['actual'] == pred['predicted'] else "✗"
            logger.info(f"  {result} Actual: {pred['actual']}, Predicted: {pred['predicted']}, Confidence: {pred['confidence']:.3f}")
    else:
        accuracy = 0.0
        logger.warning("No successful predictions made")
    
    return accuracy, train_model

def save_model_and_results(model, inference_results, accuracy, data):
    logger.info("Saving model and results")
    
    # Save the trained model
    with open('bayesian_network_model.pkl', 'wb') as f:
        pickle.dump(model, f)
    logger.info("Saved trained model as 'bayesian_network_model.pkl'")
    
    # Save inference results
    with open('inference_results.json', 'w') as f:
        json.dump(inference_results, f, indent=2)
    logger.info("Saved inference results as 'inference_results.json'")
    
    # Save processed data
    data.to_csv('processed_iris_data.csv', index=False)
    logger.info("Saved processed data as 'processed_iris_data.csv'")
    
    # Save model summary
    summary = {
        'model_type': 'Discrete Bayesian Network',
        'dataset': 'Iris (discretized)',
        'nodes': list(model.nodes()),
        'edges': list(model.edges()),
        'accuracy': accuracy,
        'data_shape': list(data.shape),
        'num_parameters': sum(cpd.values.size for cpd in model.get_cpds()),
        'inference_scenarios': len(inference_results)
    }
    
    with open('model_summary.json', 'w') as f:
        json.dump(summary, f, indent=2)
    logger.info("Saved model summary as 'model_summary.json'")

def main():
    logger.info("Starting Bayesian Network implementation")
    
    # Load and prepare data
    data = load_and_prepare_data()
    
    # Create network structure
    model = create_network_structure()
    
    # Learn parameters
    trained_model = learn_parameters(model, data)
    
    # Define inference scenarios
    evidence_scenarios = {
        'large_petals': {'Petal_Length': 2, 'Petal_Width': 2},  # Large petals
        'small_petals': {'Petal_Length': 0, 'Petal_Width': 0},  # Small petals
        'medium_sepals': {'Sepal_Length': 1, 'Sepal_Width': 1}, # Medium sepals
        'mixed_features': {'Sepal_Length': 2, 'Petal_Length': 0} # Mixed evidence
    }
    
    # Perform inference
    inference_results = perform_inference(trained_model, evidence_scenarios)
    
    # Evaluate model
    accuracy, _ = evaluate_model_performance(trained_model, data)
    
    # Create visualizations
    visualize_network_and_results(trained_model, inference_results, data)
    
    # Save everything
    save_model_and_results(trained_model, inference_results, accuracy, data)
    
    logger.info("Bayesian Network implementation completed successfully!")
    logger.info(f"Final model accuracy: {accuracy:.4f}")
    logger.info("All artifacts saved for future reference")

if __name__ == "__main__":
    main()
```

```python
# Bayesian Networks Implementation - Fully Commented Version
# This implementation demonstrates probabilistic graphical models for classification
# using the Iris dataset with discrete features

# Installation commands for dependency management
# Google Colab: !pip install pgmpy pandas numpy matplotlib seaborn scikit-learn
# Local Mac: pip install pgmpy pandas numpy matplotlib seaborn scikit-learn

import logging
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pgmpy.models import DiscreteBayesianNetwork  # Updated from deprecated BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination  # For probabilistic queries
from pgmpy.estimators import MaximumLikelihoodEstimator, BayesianEstimator
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer  # Convert continuous to discrete
import json
import pickle
import warnings
warnings.filterwarnings('ignore')  # Suppress non-critical warnings

# Configure comprehensive logging system
# This captures all major steps for debugging and analysis
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('bayesian_network_training.log'),  # Save to file
        logging.StreamHandler()  # Display in console
    ]
)
logger = logging.getLogger(__name__)

def load_and_prepare_data():
    """
    Load Iris dataset and convert continuous features to discrete bins.
    
    Bayesian Networks require discrete variables, so we discretize continuous
    features into 3 bins using quantile-based binning for balanced distribution.
    
    Returns:
        pd.DataFrame: Processed dataset with discrete features
    """
    logger.info("Loading and preparing Iris dataset for Bayesian Network")
    
    # Load the classic Iris dataset (150 samples, 4 features, 3 species)
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    
    logger.info(f"Original dataset shape: {df.shape}")
    logger.info(f"Features: {list(df.columns)}")
    
    # Discretize continuous features into 3 bins (small, medium, large)
    # Strategy='quantile' ensures each bin has approximately equal samples
    discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
    
    # Create meaningful, shorter feature names for network clarity
    feature_mapping = {
        'sepal length (cm)': 'Sepal_Length',
        'sepal width (cm)': 'Sepal_Width', 
        'petal length (cm)': 'Petal_Length',
        'petal width (cm)': 'Petal_Width'
    }
    
    df_processed = df.copy()
    
    # Transform each continuous feature to discrete bins (0, 1, 2)
    for old_name, new_name in feature_mapping.items():
        # Fit discretizer and transform data
        discretized_feature = discretizer.fit_transform(df[[old_name]]).flatten()
        # Convert to integers for cleaner representation
        df_processed[new_name] = discretized_feature.astype(int)
        logger.info(f"Discretized {old_name} -> {new_name}: {len(np.unique(discretized_feature))} bins")
    
    # Create final dataset with only the processed features we need
    final_df = df_processed[['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'species']].copy()
    final_df.rename(columns={'species': 'Species'}, inplace=True)
    
    logger.info(f"Final processed dataset shape: {final_df.shape}")
    logger.info(f"Species distribution: {final_df['Species'].value_counts().to_dict()}")
    
    return final_df

def create_network_structure():
    """
    Define the Bayesian Network structure based on botanical domain knowledge.
    
    Network Design Rationale:
    - Species is the root cause (influences all physical characteristics)
    - All morphological features depend on species
    - Petal dimensions are correlated (petal length influences petal width)
    
    This creates a DAG (Directed Acyclic Graph) representing causal relationships.
    
    Returns:
        DiscreteBayesianNetwork: Network structure without learned parameters
    """
    logger.info("Creating Bayesian Network structure")
    
    # Define edges representing causal relationships
    # Format: (parent, child) - parent causes/influences child
    model = DiscreteBayesianNetwork([
        ('Species', 'Sepal_Length'),      # Species determines sepal length
        ('Species', 'Sepal_Width'),       # Species determines sepal width
        ('Species', 'Petal_Length'),      # Species determines petal length
        ('Species', 'Petal_Width'),       # Species determines petal width
        ('Petal_Length', 'Petal_Width')   # Petal dimensions are correlated
    ])
    
    # Log network structure for verification
    logger.info(f"Network nodes: {model.nodes()}")
    logger.info(f"Network edges: {model.edges()}")
    logger.info("Network structure created successfully (DAG validation will occur after parameter learning)")
    
    return model

def learn_parameters(model, data):
    """
    Learn Conditional Probability Distributions (CPDs) from data using MLE.
    
    Maximum Likelihood Estimation finds parameter values that maximize
    the likelihood of observing the training data.
    
    Args:
        model: Network structure
        data: Training data
    
    Returns:
        DiscreteBayesianNetwork: Trained model with learned CPDs
    """
    logger.info("Learning network parameters using Maximum Likelihood Estimation")
    
    # Fit model parameters to data using Maximum Likelihood Estimation
    # This automatically creates CPDs for each node based on observed frequencies
    model.fit(data, estimator=MaximumLikelihoodEstimator)
    
    # Validate that the model is properly constructed
    try:
        model.check_model()  # Verifies all CPDs are valid and consistent
        logger.info("Model validation passed - all CPDs are properly defined")
    except Exception as e:
        logger.warning(f"Model validation warning: {e}")
    
    # Log details about learned probability distributions
    logger.info("Learned Conditional Probability Distributions:")
    for cpd in model.get_cpds():
        logger.info(f"CPD for {cpd.variable}:")
        logger.info(f"  Variables: {cpd.variables}")  # Shows variable and its parents
        
        # Safely access cardinality information (number of states per variable)
        try:
            if hasattr(cpd, 'cardinality'):
                logger.info(f"  Cardinality: {cpd.cardinality}")
            elif hasattr(cpd, 'get_cardinality'):
                logger.info(f"  Cardinality: {cpd.get_cardinality()}")
            else:
                logger.info(f"  Shape info: {cpd.values.shape}")
        except Exception as e:
            logger.info(f"  Shape info: {cpd.values.shape}")
        
        logger.info(f"  Values shape: {cpd.values.shape}")
        
        # Show sample probability values for smaller CPDs
        if cpd.values.size <= 20:  # Only for manageable sizes
            logger.info(f"  Sample values: {cpd.values.flatten()[:5]}")
    
    return model

def perform_inference(model, evidence_scenarios):
    """
    Perform probabilistic inference for different evidence scenarios.
    
    Uses Variable Elimination algorithm for exact inference to compute
    P(Species | Evidence) for various combinations of observed features.
    
    Args:
        model: Trained Bayesian Network
        evidence_scenarios: Dict of scenario_name -> evidence_dict
    
    Returns:
        dict: Results for each inference scenario
    """
    logger.info("Performing probabilistic inference")
    
    # Create inference engine using Variable Elimination (exact inference)
    inference = VariableElimination(model)
    results = {}
    
    # Process each evidence scenario
    for scenario_name, evidence in evidence_scenarios.items():
        logger.info(f"Inference scenario: {scenario_name}")
        logger.info(f"Evidence: {evidence}")
        
        try:
            # Query: Given evidence, what's the probability distribution over Species?
            result = inference.query(variables=['Species'], evidence=evidence)
            
            # Log human-readable results
            logger.info(f"Results for {scenario_name}:")
            species_names = ['Setosa', 'Versicolor', 'Virginica']
            for i, prob in enumerate(result.values):
                logger.info(f"  P(Species={species_names[i]}|Evidence) = {prob:.4f}")
            
            # Store results for later analysis and visualization
            results[scenario_name] = {
                'evidence': evidence,
                'probabilities': result.values.tolist(),
                'species_names': species_names
            }
        except Exception as e:
            # Handle cases where evidence is impossible/inconsistent
            logger.error(f"Error in inference for {scenario_name}: {e}")
            results[scenario_name] = {
                'evidence': evidence,
                'error': str(e)
            }
    
    return results

def visualize_network_and_results(model, inference_results, data):
    """
    Create comprehensive visualizations of the network and inference results.
    
    Generates 4 subplots:
    1. Species distribution in dataset
    2. Feature correlation heatmap
    3. Inference results comparison
    4. Network structure adjacency matrix
    
    Args:
        model: Trained network
        inference_results: Results from inference scenarios
        data: Original dataset
    """
    logger.info("Creating visualizations")
    
    # Set up matplotlib style and create subplot grid
    plt.style.use('default')
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Plot 1: Species distribution (shows balanced dataset)
    species_counts = data['Species'].value_counts().sort_index()
    axes[0, 0].bar(species_counts.index, species_counts.values, alpha=0.7, edgecolor='black')
    axes[0, 0].set_title('Species Distribution in Dataset')
    axes[0, 0].set_xlabel('Species (0=Setosa, 1=Versicolor, 2=Virginica)')
    axes[0, 0].set_ylabel('Count')
    axes[0, 0].set_xticks([0, 1, 2])
    
    # Plot 2: Feature correlation matrix (shows relationships between features)
    correlation_matrix = data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, ax=axes[0, 1], fmt='.3f')
    axes[0, 1].set_title('Feature Correlation Matrix')
    
    # Plot 3: Inference results comparison (shows model predictions)
    valid_results = {k: v for k, v in inference_results.items() if 'probabilities' in v}
    if valid_results:
        scenarios = list(valid_results.keys())[:3]  # Show first 3 valid scenarios
        species_names = ['Setosa', 'Versicolor', 'Virginica']
        x = np.arange(len(species_names))  # Species positions
        width = 0.25  # Bar width for grouped bars
        
        # Use distinct colors for each scenario
        colors = ['skyblue', 'lightcoral', 'lightgreen']
        for i, scenario in enumerate(scenarios):
            probs = valid_results[scenario]['probabilities']
            # Offset bars for each scenario to create grouped bar chart
            axes[1, 0].bar(x + i*width, probs, width, label=scenario, alpha=0.8, color=colors[i])
        
        axes[1, 0].set_xlabel('Species')
        axes[1, 0].set_ylabel('Probability')
        axes[1, 0].set_title('Inference Results Comparison')
        axes[1, 0].set_xticks(x + width)  # Center ticks between grouped bars
        axes[1, 0].set_xticklabels(species_names)
        axes[1, 0].legend()
        axes[1, 0].set_ylim(0, 1)  # Probability range [0, 1]
    else:
        # Handle case where no valid inference results exist
        axes[1, 0].text(0.5, 0.5, 'No valid inference results', 
                       ha='center', va='center', transform=axes[1, 0].transAxes)
        axes[1, 0].set_title('Inference Results (No Valid Results)')
    
    # Plot 4: Network structure as adjacency matrix
    nodes = list(model.nodes())
    adj_matrix = np.zeros((len(nodes), len(nodes)))  # Initialize empty matrix
    node_to_idx = {node: i for i, node in enumerate(nodes)}  # Map node names to indices
    
    # Fill adjacency matrix: 1 if edge exists, 0 otherwise
    for edge in model.edges():
        parent_idx, child_idx = node_to_idx[edge[0]], node_to_idx[edge[1]]
        adj_matrix[parent_idx][child_idx] = 1
    
    # Create heatmap of network structure
    sns.heatmap(adj_matrix, annot=True, xticklabels=nodes, yticklabels=nodes, 
                cmap='Blues', ax=axes[1, 1], cbar=False, fmt='g')
    axes[1, 1].set_title('Network Structure (Adjacency Matrix)')
    axes[1, 1].set_xlabel('Child Nodes')
    axes[1, 1].set_ylabel('Parent Nodes')
    
    # Save and display the complete visualization
    plt.tight_layout()
    plt.savefig('bayesian_network_analysis.png', dpi=300, bbox_inches='tight')
    logger.info("Saved visualization as 'bayesian_network_analysis.png'")
    plt.show()

def evaluate_model_performance(model, data):
    """
    Evaluate model performance using train/test split with stratification.
    
    Performance Evaluation Strategy:
    - Stratified split maintains class balance in train/test sets
    - Use all features except Species as evidence for prediction
    - Predict Species and compare with ground truth
    - Calculate accuracy and log sample predictions with confidence
    
    Args:
        model: Trained Bayesian Network
        data: Complete dataset
    
    Returns:
        tuple: (accuracy_score, trained_model_on_train_split)
    """
    logger.info("Evaluating model performance")
    
    # Create stratified train/test split (maintains class distribution)
    # test_size=0.3 means 30% for testing, 70% for training
    # stratify ensures equal representation of each species in both sets
    train_data, test_data = train_test_split(data, test_size=0.3, random_state=42, stratify=data['Species'])
    logger.info(f"Training data shape: {train_data.shape}")
    logger.info(f"Test data shape: {test_data.shape}")
    
    # Train a new model on training data only (proper evaluation)
    train_model = DiscreteBayesianNetwork(model.edges())  # Same structure
    train_model.fit(train_data, estimator=MaximumLikelihoodEstimator)
    
    # Create inference engine for the trained model
    inference = VariableElimination(train_model)
    correct_predictions = 0
    total_predictions = 0
    
    prediction_details = []  # Store details for analysis
    
    # Evaluate on each test sample
    for idx, row in test_data.iterrows():
        try:
            # Create evidence dictionary (all features except target Species)
            evidence = {col: int(row[col]) for col in data.columns if col != 'Species'}
            
            # Perform inference: P(Species | evidence)
            result = inference.query(variables=['Species'], evidence=evidence)
            
            # Make prediction (species with highest probability)
            predicted_species = np.argmax(result.values)
            actual_species = int(row['Species'])
            
            # Store prediction details for analysis
            prediction_details.append({
                'actual': actual_species,
                'predicted': predicted_species,
                'confidence': float(np.max(result.values)),  # Highest probability
                'evidence': evidence
            })
            
            # Count correct predictions
            if predicted_species == actual_species:
                correct_predictions += 1
            total_predictions += 1
            
        except Exception as e:
            # Log failed predictions (e.g., due to unseen evidence combinations)
            logger.warning(f"Could not make prediction for row {idx}: {e}")
    
    # Calculate and report accuracy
    if total_predictions > 0:
        accuracy = correct_predictions / total_predictions
        logger.info(f"Model accuracy on test data: {accuracy:.4f} ({correct_predictions}/{total_predictions})")
        
        # Show sample predictions with confidence scores
        logger.info("Sample predictions:")
        for i, pred in enumerate(prediction_details[:5]):  # Show first 5 predictions
            result_symbol = "✓" if pred['actual'] == pred['predicted'] else "✗"
            logger.info(f"  {result_symbol} Actual: {pred['actual']}, Predicted: {pred['predicted']}, Confidence: {pred['confidence']:.3f}")
    else:
        accuracy = 0.0
        logger.warning("No successful predictions made")
    
    return accuracy, train_model

def save_model_and_results(model, inference_results, accuracy, data):
    """
    Save all important artifacts for reproducibility and future use.
    
    Saves:
    - Trained model (pickle format for Python)
    - Inference results (JSON for cross-platform compatibility)
    - Processed data (CSV for analysis in other tools)
    - Model summary (JSON with metadata)
    
    Args:
        model: Trained Bayesian Network
        inference_results: Dictionary of inference scenarios and results
        accuracy: Model accuracy score
        data: Processed dataset
    """
    logger.info("Saving model and results")
    
    # Save trained model using pickle (preserves Python objects)
    with open('bayesian_network_model.pkl', 'wb') as f:
        pickle.dump(model, f)
    logger.info("Saved trained model as 'bayesian_network_model.pkl'")
    
    # Save inference results as JSON (human-readable, cross-platform)
    with open('inference_results.json', 'w') as f:
        json.dump(inference_results, f, indent=2)
    logger.info("Saved inference results as 'inference_results.json'")
    
    # Save processed data as CSV (can be opened in Excel, loaded by other tools)
    data.to_csv('processed_iris_data.csv', index=False)
    logger.info("Saved processed data as 'processed_iris_data.csv'")
    
    # Create comprehensive model summary
    summary = {
        'model_type': 'Discrete Bayesian Network',
        'dataset': 'Iris (discretized)',
        'nodes': list(model.nodes()),
        'edges': list(model.edges()),
        'accuracy': accuracy,
        'data_shape': list(data.shape),
        'num_parameters': sum(cpd.values.size for cpd in model.get_cpds()),  # Total parameters
        'inference_scenarios': len(inference_results)
    }
    
    # Save model summary as JSON
    with open('model_summary.json', 'w') as f:
        json.dump(summary, f, indent=2)
    logger.info("Saved model summary as 'model_summary.json'")

def main():
    """
    Main execution function that orchestrates the complete Bayesian Network workflow.
    
    Workflow:
    1. Data loading and preprocessing
    2. Network structure definition
    3. Parameter learning from data
    4. Probabilistic inference with test scenarios
    5. Model evaluation and performance measurement
    6. Visualization and results analysis
    7. Artifact saving for reproducibility
    """
    logger.info("Starting Bayesian Network implementation")
    
    # Step 1: Load and preprocess data
    data = load_and_prepare_data()
    
    # Step 2: Define network structure based on domain knowledge
    model = create_network_structure()
    
    # Step 3: Learn parameters (CPDs) from data using MLE
    trained_model = learn_parameters(model, data)
    
    # Step 4: Define test scenarios for inference
    # These scenarios test different combinations of evidence
    evidence_scenarios = {
        'large_petals': {'Petal_Length': 2, 'Petal_Width': 2},    # Expect Virginica
        'small_petals': {'Petal_Length': 0, 'Petal_Width': 0},   # Expect Setosa  
        'medium_sepals': {'Sepal_Length': 1, 'Sepal_Width': 1},  # Mixed prediction
        'mixed_features': {'Sepal_Length': 2, 'Petal_Length': 0} # Potentially conflicting
    }
    
    # Step 5: Perform probabilistic inference
    inference_results = perform_inference(trained_model, evidence_scenarios)
    
    # Step 6: Evaluate model performance using proper train/test methodology
    accuracy, _ = evaluate_model_performance(trained_model, data)
    
    # Step 7: Create comprehensive visualizations
    visualize_network_and_results(trained_model, inference_results, data)
    
    # Step 8: Save all artifacts for future reference and reproducibility
    save_model_and_results(trained_model, inference_results, accuracy, data)
    
    # Final summary
    logger.info("Bayesian Network implementation completed successfully!")
    logger.info(f"Final model accuracy: {accuracy:.4f}")
    logger.info("All artifacts saved for future reference")

# Entry point - execute main function when script is run directly
if __name__ == "__main__":
    main()
```
